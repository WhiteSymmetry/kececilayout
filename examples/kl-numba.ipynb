```python
import kececilayout as kl
import networkx as nx
import matplotlib.pyplot as plt
import numba

# Örnek bir graf oluştur
G = nx.Graph()
G.add_nodes_from(range(10))  # 0'dan 9'a düğümler

# Koordinatları Hesaplama
# Parametreler
primary_spacing = 1.0
secondary_spacing = 0.5
primary_direction = 'left-to-right'
secondary_start = 'up'
expanding = True

# Koordinatları hesapla
pos = kl.calculate_coordinates(
    nodes=list(G.nodes()),
    primary_spacing=primary_spacing,
    secondary_spacing=secondary_spacing,
    primary_direction=primary_direction,
    secondary_start=secondary_start,
    expanding=expanding
)

# Sonuçları yazdır
for node, coord in pos.items():
    print(f"Düğüm {node}: Koordinatlar {coord}")

# Grafı çiz
ax = kl.draw_kececi(
    G,
    pos=pos,  # Önceden hesaplanmış koordinatlar
    style='default',
    with_labels=True,
    node_color='lightblue',
    node_size=500,
    font_weight='bold'
)

plt.title("Keçeci Layout: Sıralı-Zigzag Düzen")
plt.show()
```
